```python
import tweepy
import pandas as pd
import csv
from more_itertools import unique_everseen
```

```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
user = api.me()
print('Name: ' + user.name)
print('Location: ' + user.location)
print('Friends: ' + str(user.friends_count))
```

```python
# Open/create a file to append data to
csvFile = open('action_keyword.csv', 'a', encoding='utf-8')
#Use csv writer
csvWriter = csv.writer(csvFile)
for tweet in tweepy.Cursor(api.search,
                           q = "Action Movie",
                           since = "2019-01-01",
                           until = "2019-04-20",
                           #geocode = "39.7837304, -100.4458825, 2500km",
                           result_type='recent',
                           include_entities=True,
                           monitor_rate_limit=True, 
                           wait_on_rate_limit=True,
                           lang = "en").items():
    if not tweet.retweeted:
    # Write a row to the CSV file. I use encode UTF-8
        csvWriter.writerow([tweet.created_at, tweet.text.encode('utf-8'), tweet.user.location])
            #print (tweet.created_at, tweet.text, tweet.user.location)
csvFile.close()
```

```python
df = pd.read_csv('action_keyword.csv')
df.to_csv('action_keyword_2.csv', index=False)
```

```python
f2 = open('C:/Users/Vaidehi/action_keyword_2.csv', 'r', encoding='utf-8')
reader2 = csv.reader(f2)
f3 = open('C:/Users/Vaidehi/action_keyword_3.csv', 'a', encoding='utf-8')
writer = csv.writer(f3)

for row in reader2:
    if ("USA" in row[2]):
        writer.writerow(row)
    else:
        ff4 = open('C:/Users/Vaidehi/us_states.csv', 'r', encoding='utf-8')
        reader = csv.reader(ff4) 
        for rrrr in reader:
            if (rrrr[0].lower() in row[2].lower()):
                writer.writerow(row)
                break
            else:
                if(row[2].endswith(rrrr[1])):
                    writer.writerow(row)
                    break
                    
f2.close()
f3.close()
```
